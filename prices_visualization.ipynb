```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
amd = pd.read_excel('data/amd (2).xlsx')
nvidia = pd.read_excel('data/nvidia.xlsx')
pd.to_datetime(amd['Дата'], format='%Y-%m-%d')
pd.to_datetime(nvidia['Дата'], format='%Y-%m-%d')
```

```python
amd.info()
```

```python
nvidia.info()
```

```python
amd.sort_values(by='Дата', inplace=True, ascending=True)
amd.head(10)
```

```python
nvidia.sort_values(by='Дата', inplace=True, ascending=True)
nvidia.head()
```

```python
amd.plot(x='Дата', y='Цена после', label='Цены акций компании AMD')
plt.show()
```

```python
nvidia.plot(x='Дата', y='Цена после', label='Цены акций компании Nvidia')
```
